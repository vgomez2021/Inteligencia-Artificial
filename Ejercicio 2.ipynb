```python
import random
from fuzzywuzzy import fuzz

# 1. Lista de caracteres del abecedario
alphabet = list('abcdefghijklmnopqrstuvwxyz')

# 2. Obtener la palabra a adivinar
def get_word():
    while True:
        word = input("Ingrese una palabra de cuatro letras: ")
        if len(word) == 4:
            return word.lower()
        else:
            print("La palabra debe tener exactamente 4 letras.")

# 3. Crear una población inicial aleatoria
def create_population(size):
    population = []
    for _ in range(size):
        individual = ''.join(random.choice(alphabet) for _ in range(4))
        population.append(individual)
    return population

# 4. Calcular el fitness de un individuo
def fitness(individual, target):
    return fuzz.ratio(individual, target)

# 5. Seleccionar padres para la reproducción
def selection(population, fitnesses):
    # Selección por ruleta
    total_fitness = sum(fitnesses)
    probabilities = [f / total_fitness for f in fitnesses]
    return random.choices(population, weights=probabilities, k=2)

# 6. Cruzar a los padres para crear hijos
def crossover(parent1, parent2):
    crossover_point = random.randint(1, 3)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

# 7. Mutar a los hijos
def mutation(individual, mutation_rate):
    new_individual = list(individual)
    for i in range(len(new_individual)):
        if random.random() < mutation_rate:
            new_individual[i] = random.choice(alphabet)
    return ''.join(new_individual)

# 8. Algoritmo genético principal
def genetic_algorithm(target, population_size, generations, mutation_rate):
    population = create_population(population_size)
    for generation in range(generations):
        fitnesses = [fitness(individual, target) for individual in population]
        new_population = []
        while len(new_population) < population_size:
            parent1, parent2 = selection(population, fitnesses)
            child1, child2 = crossover(parent1, parent2)
            child1 = mutation(child1, mutation_rate)
            child2 = mutation(child2, mutation_rate)
            new_population.extend([child1, child2])
        population = new_population
        best_individual = max(population, key=lambda x: fitness(x, target))
        print(f"Generación {generation}: {best_individual} ({fitness(best_individual, target)})")
        if fitness(best_individual, target) == 100:
            return best_individual

# Obtener la palabra a adivinar
target_word = get_word()

# Ejecutar el algoritmo genético
result = genetic_algorithm(target_word, population_size=100, generations=100, mutation_rate=0.01)

print(f"La palabra adivinada es: {result}")



```
