```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Importar el archivo
df = pd.read_csv('nba.csv')

# 2. Modificar el campo nombre por jugador
df.rename(columns={'nombre': 'jugador'}, inplace=True)

# 3. Menú de opciones para el usuario
while True:
    print("¿Qué deseas hacer?")
    print("1. Ver todos los jugadores")
    print("2. Ver detalle por jugador")
    print("3. Ver el puntaje general")
    print("4. Graficar")
    print("5. Salir")
    opcion = input("Ingrese una opción: ")

    if opcion == '1':
        # Ver todos los jugadores
        print(df['jugador'])
    elif opcion == '2':
        # Ver detalle por jugador
        jugador = input("Ingrese el nombre del jugador: ")
        print(df[df['jugador'] == jugador])
    elif opcion == '3':
        # Ver el puntaje general
        print("El puntaje promedio es:", df['puntos'].mean())
    elif opcion == '4':
        # Graficar
        # Manejar nombres duplicados
        df_copy = df.copy()
        df_copy['jugador'] = df_copy['jugador'].astype(str) + df_copy.index.astype(str)  # Agregar un índice único
        plt.figure(figsize=(10, 10))
        plt.bar(df_copy['jugador'], df_copy['puntos'], color='green')
        plt.xlabel('Jugador')
        plt.ylabel('Puntos')
        plt.title('Puntos por jugador')
        plt.xticks(rotation=90)
        plt.show()
    elif opcion == '5':
        # Salir
        break
    else:
        print("Opción inválida. Por favor, ingrese una opción válida.")
```
